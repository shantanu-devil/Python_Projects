```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

## 1. Scatter Plots:

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

plt.scatter(x_data, y_data, c = "red", marker = "*", s = 30, alpha = 0.7)
plt.title('Random Points', fontsize=14)
plt.xlabel('X-Axis', fontsize=12)
plt.ylabel('Y-Axis', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```

## 2. Line Plots:

```python
x = [1, 2, 3, 4, 5]  
y = [2, 3, 5, 7, 11] 

plt.plot(x, y, color='green', linestyle='--', marker='o', linewidth=2, markersize=8, label='Data Line')
plt.title('Basic Line Plot') 
plt.xlabel('X Axis')  
plt.ylabel('Y Axis')  
plt.legend() 
plt.grid(True)  
plt.show()
```

```python
data = {'X': [1, 2, 3, 4, 5],
        'Y': [2, 3, 5, 7, 11]}
df = pd.DataFrame(data)

df.plot(x='X', y='Y', kind='line', title='Line Plot from DataFrame')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

y2 = [1, 4, 6, 8, 10]

plt.plot(x, y, label='Line 1')
plt.plot(x, y2, label='Line 2', linestyle=':', marker='^')

plt.title('Multiple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)

plt.show()
```

## 3. Bar Plots:

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 78]


plt.bar(categories, values, color=['red', 'green', 'blue', 'cyan'], width=0.5)


plt.title('Basic Vertical Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

plt.barh(categories, values)
plt.title('Basic Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
df = pd.DataFrame({'Values': values}, index=categories)


df.plot(kind='bar', legend=False)


plt.title('Bar Plot from DataFrame')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

## 4. Histograms:

```python
data = np.random.randn(1000) 

plt.hist(data, bins=30, edgecolor='black', density=True, alpha=0.6, color ='orange', cumulative=False)


plt.title('Histogram of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normalized Histogram with Custom Colors')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```

```python
df = pd.DataFrame({'Data': data})

df['Data'].plot(kind='hist', bins=5, edgecolor='black', alpha=0.7)

plt.title('Histogram from Pandas DataFrame')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

## 5. Pie Charts:

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]  
explode = (0, 0.1, 0, 0) 


plt.pie(sizes, explode=explode, labels=labels, colors=['lightblue', 'lightgreen', 'lightpink', 'lightyellow'], autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  
plt.title('Customized Pie Chart')
plt.show()
```

## 6. Boxplots:

```python
data = np.random.randn(100) 

plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')

plt.show()

data1 = np.random.normal(100, 10, 200)  
data2 = np.random.normal(80, 30, 200)   
data3 = np.random.normal(90, 20, 200)   

data = [data1, data2, data3]

plt.boxplot(data, patch_artist=True, notch=True, labels=['Group 1', 'Group 2', 'Group 3'],
            medianprops={'linewidth': 2, 'color': 'purple'},
            boxprops={'facecolor': 'lightblue', 'edgecolor': 'black'},
            whiskerprops={'color': 'green'},
            capprops={'color': 'red'},
            flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 5, 'linestyle': 'none'})

plt.title('Customized Box Plot')
plt.ylabel('Values')

plt.show()
```

## 7. Plot Customization:

```python
x = np.linspace(0, 10, 100)


plt.plot(x, np.sin(x), color='blue', linewidth=2, linestyle='-', label='sin(x)')
plt.plot(x, np.cos(x), color='red', linewidth=1, linestyle='--', label='cos(x)')
plt.xlim([0, 10])
plt.ylim([-1.5, 1.5])
plt.legend()
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 2, 0.5))
plt.grid(True)
plt.figure(figsize=(10, 6), dpi=80)
plt.savefig('my_plot.png', dpi=300)
plt.style.use('ggplot') 
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4)) 

ax[0].plot(x, np.sin(x), color='blue')
ax[0].set_title('Sine Curve')

ax[1].plot(x, np.cos(x), color='red')
ax[1].set_title('Cosine Curve')
```

## 8. 3D Plotting:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')

ax.plot_surface(x, y, z, cmap='viridis')
```

## 9. Animating Plots:

```python
fig, ax = plt.subplots()

x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def init():
    line.set_ydata(np.sin(x))
    return line,

def update(frame):
    line.set_ydata(np.sin(x + frame / 10.0))
    return line,


ani = FuncAnimation(fig, update, frames=range(200), init_func=init, blit=True)

plt.show()
```
