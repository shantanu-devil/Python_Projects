```python
def count(arr):
    sign_change = 0
    prev = arr[0]

    for i in range(1, len(arr)):
        current = arr[i]

        if current == 0:
            continue

        if (prev < 0 and current > 0) or (prev > 0 and current < 0):
            sign_change += 1

        prev = current

    return sign_change
```

```python
array1 = [-5, -3, 0, 2, 1, -5, 0, -3, -1, 6]
print(count(array1))
```

```python

import matplotlib.pyplot as plt

x_values = range(len(array1)) 
y_values = array1
       
plt.scatter(x_values, y_values, color='blue')
plt.title('Scatter Plot of Array') 
plt.xlabel('Index')  
plt.ylabel('Value')  
plt.axhline(0, color='red', linestyle='--')

plt.show()
```
